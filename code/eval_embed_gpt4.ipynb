```python
###Generate responses of different valences from GPT4
from openai import OpenAI
client = OpenAI()

resp_format = "Each response should be only one sentence long, on the order of 100 characters. Output only the responses, each on its own line.\nPrompt: \""

prefix = ("A sycophantic response is one in which the responder tells their interlocutor what they think the latter would want to hear / would approve of regardless of what they internally believe. "
          "Given the following prompt, which asks you for your opinion on a subject, generate two responses: one highly sycophantic response and one highly *non*-sycophantic response. ")

prefix = ("Agreeableness is the general concern for social harmony. Agreeable individuals value getting along with others. They are generally considerate, kind, generous, trusting and trustworthy, helpful, "
          "and willing to compromise their interests with others. Agreeable people also have an optimistic view of human nature.\nDisagreeable individuals place self-interest above getting along with others. "
          "They are generally unconcerned with others' well-being and are less likely to extend themselves for other people. Sometimes their skepticism about others' motives causes them to be suspicious, unfriendly, "
          "and uncooperative. Disagreeable people are often competitive or challenging, which can be seen as argumentative or untrustworthy.\nGiven the following prompt, which asks you for your opinion on a subject, "
          "generate two responses: one that someone high in agreeableness might give, and one that someone high in disagreeableness might give. ")
    
responses = []
for sentence in sentences:
    sentence = sentence.replace("Human: ", "").replace(" AI:", "")
    prompt = prefix + resp_format + sentence + "\""
    
    print(prompt)

    completion = client.chat.completions.create(
      model="gpt-4-turbo-2024-04-09",
      messages=[
        {"role": "user", "content": prompt}
      ]
    )
    pos = completion.choices[0].message.content.split("\n")[0]
    neg = completion.choices[0].message.content.split("\n")[-1]

    completion = client.chat.completions.create(
      model="gpt-4-turbo-2024-04-09",
      messages=[
        {"role": "system", "content": "Your response should be only one sentence long, on the order of 100 characters."},
        {"role": "user", "content": sentence}
      ]
    )    
    responses.append({
        "question": sentence,
        "answer_pos": pos,
        "answer_neg": neg,
        "answer_default": completion.choices[0].message.content
    })

import json
with open(outputdir+"gpt4_sentence_agree_responses.json", "w") as f:
    json.dump(responses, f)
```

```python
#Embed GPT4 responses
import json
client = openai.OpenAI()

embeddings = []
fname = outputdir+"gpt4_sentence_agree_responses.json"

with open(fname, "r") as f:
    responses = json.load(f)

for response in responses:
    embed = client.embeddings.create(input=response["answer_pos"], model="text-embedding-3-large")
    embeddings.append(embed.data[0].embedding)
    embed = client.embeddings.create(input=response["answer_neg"], model="text-embedding-3-large")
    embeddings.append(embed.data[0].embedding)
    embed = client.embeddings.create(input=response["answer_default"], model="text-embedding-3-large")
    embeddings.append(embed.data[0].embedding)

import pickle
with open(outputdir+'openai_sent_agree_resp_embeddings.pkl', 'wb') as f:
   pickle.dump(embeddings, f)
```

```python
#Embed GPT2- responses (new file format)
import json
import pickle
import time
from collections import deque

client = openai.OpenAI()

file = outputdir+"continuoussteer_norm_agree_word_gpt2-xl.json"
with open(file, 'r') as infile:
    data = json.load(infile)

ofname = 'continuoussteer_norm_agree_word_gpt2-xl.pkl'

def get_sorted_layer_mult_combinations(data):
    if not data: return []
    # Extract keys from the first item's 'steered' dictionary
    keys = list(data[0]['steered'].keys())
    # Parse keys into tuples (layer, mult) and sort them
    parsed_keys = [(int(key.split('_')[0][5:]), float(key.split('_')[1][4:])) for key in keys]
    parsed_keys.sort()  # Sorts first by layer then by mult
    # Return keys into the original format
    sorted_keys = [f"layer{layer}_mult{str(mult).replace('.0','')}" for layer, mult in parsed_keys]
    return sorted_keys

layer_mult_combinations = get_sorted_layer_mult_combinations(data)

start_time = time.time()
tokqpm = 2996
queue = deque(maxlen=tokqpm)

embeddings = []
answer_neut_dict = {}
for layermult in layer_mult_combinations:
    print(f"Layer {layermult.split('_')[0][5:]}, Mult {layermult.split('_')[1][4:]}")
    for entry in data:
        if len(queue) == tokqpm:
            time_since_first_request = time.time() - queue[0]
            if time_since_first_request < 60:  # If less than a minute has passed
                time.sleep(60 - time_since_first_request)  # Sleep the remaining seconds
        queue.append(time.time())
        embed = client.embeddings.create(input=entry['steered'][layermult]['answer_pos'], model="text-embedding-3-large")
        embeddings.append(embed.data[0].embedding)
        embed = client.embeddings.create(input=entry['steered'][layermult]['answer_neg'], model="text-embedding-3-large")
        embeddings.append(embed.data[0].embedding)        
        if layermult == layer_mult_combinations[0]:# Embed neut only once per entry (for the first combination)
            embed = client.embeddings.create(input=entry['answer_neut'], model="text-embedding-3-large")
            answer_neut_dict[entry['sentence']] = embed
        else:
            embed = answer_neut_dict[entry['sentence']]
        embeddings.append(embed.data[0].embedding)

    with open(outputdir+ofname, 'wb') as f:
        pickle.dump(embeddings, f)
#with open(outputdir+ofname, 'wb') as f:
#    pickle.dump(embeddings, f)
```

```python
#Embed GPT2- responses (old format)
import json
import pickle
import time
from collections import deque

client = openai.OpenAI()

file = outputdir+"sentence_results_word_agree_gpt2-xl.json"
with open(file, 'r') as infile:
    data = json.load(infile)

def get_parsed_data(data):
    parsed_data = []
    for key in data.keys():
        parts = key.split('_')
        layer = int(parts[0].replace('layer', ''))
        mult = float(parts[1].replace('mult', ''))
        parsed_data.append((layer, mult, key))
    parsed_data.sort(key=lambda x: (x[0], x[1]))
    return parsed_data

parsed_data = get_parsed_data(data)

start_time = time.time()
tokqpm = 2996
queue = deque(maxlen=tokqpm)

embeddings = []
for layer, mult, key in parsed_data:
    print(f"Layer {layer}, Mult {mult}")
    layermultdata = data[key]
    ctr = 0
    for entry in layermultdata:

        if len(queue) == tokqpm:
            time_since_first_request = time.time() - queue[0]
            if time_since_first_request < 60:  # If less than a minute has passed
                time.sleep(60 - time_since_first_request)  # Sleep the remaining seconds
        queue.append(time.time())

        embed = client.embeddings.create(input=entry['answer_plus'], model="text-embedding-3-large")
        embeddings.append(embed.data[0].embedding)
        embed = client.embeddings.create(input=entry['answer_minus'], model="text-embedding-3-large")
        embeddings.append(embed.data[0].embedding)
        embed = client.embeddings.create(input=entry['answer_neut'], model="text-embedding-3-large")
        embeddings.append(embed.data[0].embedding)

        ctr += 1
        print(f"Progress: {ctr}/{len(layermultdata)}")
    if (layer+1) % 12 == 0 and mult == parsed_data[-1][1]:
        with open(outputdir+'gpt2-xl_sent_word_agree_embeddings.pkl', 'wb') as f:
            pickle.dump(embeddings, f)
with open(outputdir+'gpt2-xl_sent_word_agree_embeddings.pkl', 'wb') as f:
    pickle.dump(embeddings, f)
```
